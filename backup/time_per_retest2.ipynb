```python
# (2) LIc積192.５の時のタイミングパラメータの比較

LIc = 192.5

# ファイルからのデータの読み込み
df_hold = pd.read_csv(hold_file, names=["lic", "Ic", "mean", "std"])
df_setup = pd.read_csv(setup_file, names=["lic", "Ic", "mean", "std"])

# LIc積に合致するデータの抽出
df_hold = df_hold[df_hold["lic"] == LIc]
x = df_hold["Ic"]
hold_mean = df_hold["mean"]
hold_std = df_hold["std"]
df_setup = df_setup[df_setup["lic"] == LIc]
setup_mean = df_setup["mean"]
setup_std = df_setup["std"]

# グラフの描画(平均)
plt.figure(figsize=(4, 5))
plt.rcParams['font.family'] = ''
# plt.title('Icとセットアップ、ホールドの平均との関係(LIc=${}$)'.format(LIc))
# plt.xlabel('臨界電流値(μA)')
# plt.ylabel('平均値(ps)')
plt.xlabel('Ic (μA)')
plt.ylabel('Mean (ps)')

# y軸の範囲を設定
plt.ylim(-6, 6)

# y軸のメモリを少なくする
y_ticks = np.arange(-10, 11, 2)  # -10から10の範囲で2単位ごと
plt.yticks(y_ticks)

# グラフに点を追加 (marker='o')
plt.plot(x, hold_mean, label="Hold Mean", marker='o', linestyle='-', color='blue')
plt.plot(x, setup_mean, label="Setup Mean", marker='x', linestyle='-', color='red')
plt.legend()
plt.show()

# グラフの描画(標準偏差)
plt.figure(figsize=(4, 5))
plt.plot(x, hold_std, label="Hold Std", marker='o', linestyle='-', color='green')
plt.plot(x, setup_std, label="Setup Std", marker='x', linestyle='-', color='purple')
# plt.title('Icとセットアップ、ホールドの標準偏差との関係(LIc=${}$)'.format(LIc))
plt.xlabel('Ic (μA)')
plt.ylabel('Std (ps)')
plt.legend()
plt.show()
```
